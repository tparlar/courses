```python
# Yaşı 30 dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz.

df[(df["age"] < 30) & (df["sex"]=="female") ]
```

```python
# Fare'i 500'den büyük veya yaşı 70 den büyük yolcuların bilgilerini gösteriniz.
df[ (df["fare"] > 500) | (df["age"] >70)]
```

```python
# who değişkenini dataframe'den düşürün.
df.drop("who",axis=1,inplace=True)
```

```python

# deck değişkenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.
```

```python
x = df["deck"].mode()[0]
df["deck"].fillna(x,inplace=True)
```

```python
df.head()
```

```python
# age değikenindeki boş değerleri age değişkenin medyanı ile doldurun.
x = df["age"].median()
df["age"].fillna(x, inplace=True)
# df["age"].fillna(df["age"].median(), inplace=True)
```

```python

# survived değişkeninin Pclass ve Cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.
df.groupby(["pclass","sex"]).agg({"survived":["sum","count","mean"]})
```

```python

# 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.
# Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz.
# (apply ve lambda yapılarını kullanınız)
# df.loc[df["age"] < 30, "age_flag"] = 1
# df.loc[df["age"] >= 30, "age_flag"] = 0
# df["age_flag"] = [ 1 if degisken < 30 else 0 for degisken in df["age"] ] ## 1. yol
df["age_flag"] = df["age"].apply(lambda x: 1 if x<30 else 0) ## 2. yol
```

```python

# Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.
tips = sns.load_dataset("tips")
tips.head()
tips.shape
```

```python
tips.value_counts()
```

```python

# Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill
# değerlerinin toplamını, min, max ve ortalamasını bulunuz.
tips.groupby(["time"]).agg({"total_bill":["min","max","mean"]})
```

```python

# Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

tips.groupby(["day","time"]).agg({"total_bill":["sum","min","max","mean"]})
```

```python

# Lunch zamanına ve kadın müşterilere ait total_bill ve tip
# değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.
tips[(tips["sex"]=="Female") & (tips["time"]=="Lunch")].groupby(["day"]).agg({"total_bill":["sum","min","max","mean"],"tip":["sum","min","max","mean"]})
```

```python

#  size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?
tips[ (tips["size"] < 3) & (tips["total_bill"] > 10)][["total_bill","tip","size"]].agg(["mean"])
```

```python

# total_bill_tip_sum adında yeni bir değişken oluşturun.
# Her bir müşterinin ödediği totalbill ve tip in toplamını versin.
tips["total_bill_tip_sum"] = tips[["total_bill","tip"]].sum(axis=1)
tips.head()
```

```python

# Total_bill değişkeninin kadın ve erkek için ayrı ayrı ortalamasını bulun.
# Bulduğunuz ortalamaların altında olanlara 0, üstünde ve eşit olanlara bir verildiği
# yeni bir total_bill_flag değişkeni oluşturun.
# Dikkat !! Female olanlar için kadınlar için bulunan ortalama dikkate alıncak, male için ise erkekler için bulunan ortalama.
# parametre olarak cinsiyet ve total_bill alan bir fonksiyon yazarak başlayın. (If-else koşulları içerecek)
```

```python
def ortalamaKontrol(gender, bill):
  female_avg = tips[tips["sex"]=="Female"]["total_bill"].mean()
  male_avg = tips[tips["sex"]=="Male"]["total_bill"].mean()
  if gender == "Female" :
    if bill >= female_avg:
      return 1
    else:
      return 0
  else:
    if bill >= male_avg:
      return 1
    else:
      return 0
```

```python

# total_bill_flag değişkenini kullanarak cinsiyetlere göre ortalamanın
# altında ve üstünde olanların sayısını gözlemleyin.
# tips.groupby(["sex","total_bill_flag"]).agg({"time":"count"})
pd.crosstab(tips["sex"],tips["total_bill_flag"])
```
