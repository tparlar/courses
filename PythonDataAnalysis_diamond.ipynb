Linear Regression ile performans karşılaştırmak için SVR deneyelim. SVR, doğrusal olmayan ilişkileri yakalama konusunda güçlüdür.

### SVR Modelini Parametrelerle Kullanma Örneği

SVR modelini oluştururken `C`, `epsilon` ve `kernel` gibi hiperparametreleri doğrudan belirleyebilirsiniz. Bu parametreler, modelin karmaşıklığını ve genelleştirme yeteneğini kontrol eder.

```python
from sklearn.svm import SVR

# SVR modelini belirli parametrelerle başlatma
svr = SVR(kernel='linear', C=10, gamma=1)

# Modeli eğitim verileriyle eğitme
svr.fit(X_train_scaler, y_train)

# Test veri seti üzerinde tahmin yapma
y_pred = svr.predict(X_test_scaler)

# R-squared skorunu hesaplama ve yazdırma
score = r2_score(y_test, y_pred)
print(f"Belirli parametrelerle SVR R-squared Score: {score}")
```

Bu örnekte:
- `kernel='rbf'`: Çekirdek fonksiyonu olarak Radyal Temel Fonksiyon (Radial Basis Function) kullanıldı. Bu, doğrusal olmayan ilişkileri modellemek için yaygın olarak kullanılır.
- `C=10`: Düzenlileştirme parametresidir. Yüksek C değeri, eğitim hatalarına daha az tolerans gösterir ve daha karmaşık bir modellemeye yol açabilir. Düşük C değeri ise daha basit bir modellemeyi teşvik eder.
- `epsilon=0.1`: Epsilon, tahmin hatasının tolere edileceği maksimum eşiktir. Model, tahminleri gerçek değerden epsilon kadar saptığında herhangi bir ceza uygulamaz.
- `gamma=0.1`: Sadece `rbf`, `poly` ve `sigmoid` çekirdekleri için bir parametredir. Yüksek gamma değeri, modelin eğitim verilerine daha yakından uymasına neden olur (overfitting riski artar); düşük gamma değeri ise daha geniş bir etki alanına sahip olup daha yumuşak bir karar sınırı oluşturur.
